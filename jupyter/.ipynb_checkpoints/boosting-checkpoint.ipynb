## Try a reasonable set of hps & Learning curve

```python
clf = AdaBoostClassifier()    # default tree is a stump, DecisionTreeClassifier(max_depth=1)
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
```

```python
title = "Learning Curves (AdaBoost)"
plot_learning_curve(clf, title, X_train, y_train, ylim=(0.7, 1.01), cv=cv, metric=roc_auc_score, n_jobs=4)

# cross validation
scores = cross_val_score(clf, X_train, y_train, scoring=make_scorer(roc_auc_score), cv=cv, n_jobs=4)
# print(scores)
print(f'mean: {scores.mean()}')
print(f'std: {scores.std()}')

# analysis
# this is underfitting
```

## Model complexity curve

```python
title = "Model Complexity Curves (AdaBoost)"
dst = DecisionTreeClassifier(max_depth=4, criterion='gini')
clf = AdaBoostClassifier(base_estimator=dst, learning_rate=1)
plot_validation_curve(clf, title, X_train, y_train, ylim=(0.7, 1.01), cv=cv, metric=roc_auc_score, n_jobs=4, param_name='n_estimators', param_range=np.arange(1, 300, 10))
```

```python
title = "Model Complexity Curves (AdaBoost)"
dst = DecisionTreeClassifier(max_depth=10, criterion='gini')
clf = AdaBoostClassifier(base_estimator=dst, learning_rate=1)
plot_validation_curve(clf, title, X_train, y_train, ylim=(0.7, 1.01), cv=cv, metric=roc_auc_score, n_jobs=4, param_name='n_estimators', param_range=np.arange(1, 300, 10))
# analysis
# 1. if base learner is more complex, boosting is tend to push the score to the limit earlier (with less number of learners)
# 2. adaboost is resistant to overfitting
```

```python
title = "Model Complexity Curves (AdaBoost)"
dst = DecisionTreeClassifier(max_depth=10, criterion='gini')
clf = AdaBoostClassifier(base_estimator=dst, learning_rate=1)
plot_validation_curve(clf, title, X_train, y_train, ylim=(0.7, 1.01), cv=cv, metric=roc_auc_score, n_jobs=4, param_name='n_estimators', param_range=np.arange(100, 400, 10))
```

## Model Selection

```python
params = {'base_estimator__max_depth': np.arange(1, 20, 1), 
          'base_estimator__criterion': ['gini'],
          'n_estimators': np.arange(40, 300, 10)
         }
dst = DecisionTreeClassifier()
clf = AdaBoostClassifier(base_estimator=dst, learning_rate=1)
gs = GridSearchCV(clf, params, scoring=make_scorer(roc_auc_score), cv=cv)
gs.fit(X_train, y_train)
# 2h
```

```python
# plot - ROC-AUC
fpr, tpr, _ = roc_curve(y_test, y_prob[:, 1])
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=1, label='ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")

# plot - PR-AUC
precision, recall, _ = precision_recall_curve(y_test, y_prob[:, 1])
pr_auc = average_precision_score(y_test, y_prob[:, 1])

plt.figure()
plt.plot(recall, precision, color='darkorange', lw=1, label='PR curve (area = %0.3f)' % pr_auc)
plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall(True Positive Rate)')
plt.ylabel('Precision')
plt.title('Precision Recall Curve')
plt.legend(loc="upper right")
```
