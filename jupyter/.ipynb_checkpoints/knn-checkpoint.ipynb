## Try a reasonable set of hps & Learning curve

```python
clf = KNeighborsClassifier(n_neighbors=5, n_jobs=4)  # default params
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)

title = "Learning Curves (KNN)"
plot_learning_curve(clf, title, X_train, y_train, ylim=(0.7, 1.01), cv=cv, metric=roc_auc_score, n_jobs=4)

# analysis
# 1. plotting the learning curve is sort of meaningless
#    training score certainly increases as the training size increases, since when training size is small, it is likely that near points are not that similiar. 
#    As more examples available, similar points start to cluster, in this case, training score increases. 
```

```python
clf = KNeighborsClassifier(n_neighbors=1, n_jobs=4)

title = "Learning Curves (KNN)"
plot_learning_curve(clf, title, X_train, y_train, ylim=(0.7, 1.01), cv=cv, metric=roc_auc_score, n_jobs=4)
# when k=1, training score is always 1
```

## Model complexity curve

```python
title = "Validation Curves (KNN)"
clf = KNeighborsClassifier(weights='uniform', n_jobs=4)
_, _ = plot_validation_curve(clf, title, X_train, y_train, ylim=(0.6, 1.01), cv=cv, metric=roc_auc_score, n_jobs=4, param_name='n_neighbors', param_range=np.arange(1, 30, 2))
# when the number of neighbors is small, ...
# when the number of neighbors is large, ...
```

```python
title = "Validation Curves (KNN)"
clf = KNeighborsClassifier()
train_score, val_score = plot_validation_curve(clf, title, X_train, y_train, ylim=(0.6, 1.01), cv=cv, metric=roc_auc_score, n_jobs=4, param_name='p', param_range=[1, 2, 3])  
# no difference when using different distance metrics
```

## Model Selection

```python
params = {'n_neighbors': np.arange(1, 15, 2), 
          'weights': ['uniform', 'distance'],
          'p': [1, 2]
         }
clf = KNeighborsClassifier(n_jobs=4)
gs = GridSearchCV(clf, params, scoring=make_scorer(roc_auc_score), cv=cv)
gs.fit(X_train, y_train)
```

```python
gs.grid_scores_
```

```python
gs.best_params_
# n_neighbors: when k=1, 1-NN this point depends only of 1 single other point
#              the model has low bias but high variance. Since we get best cross validation performance when k=1, it shows that the data is not that noisy overall. 
# p: manhattan distance
# uniform: 
```

```python
gs.best_score_
```

## Re-plot learning curve

```python
clf = KNeighborsClassifier(**gs.best_params_, n_jobs=4)
title = "Learning Curves (KNN)"
plot_learning_curve(clf, title, X_train, y_train, ylim=(0.7, 1.01), cv=cv, metric=roc_auc_score, n_jobs=4)

# cross validation
scores = cross_val_score(clf, X_train, y_train, scoring=make_scorer(roc_auc_score), cv=cv, n_jobs=4)
print(f'mean: {scores.mean()}')
print(f'std: {scores.std()}')
```
